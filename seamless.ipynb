```python
from flask import Flask, request, send_file
from flask_ngrok import run_with_ngrok
from flask_cors import CORS
import torch
import torchaudio
from transformers import SeamlessM4TModel, AutoProcessor
import os

# ✅ 1. Define correct model path
model_path = "facebook/hf-seamless-m4t-medium"

# ✅ 2. Load the SeamlessM4T model & processor
device = "cuda" if torch.cuda.is_available() else "cpu"
model = SeamlessM4TModel.from_pretrained(model_path).to(device)
processor = AutoProcessor.from_pretrained(model_path)

# ✅ 3. Setup Flask app
app = Flask(__name__)
CORS(app)  # Enable CORS for all routes
run_with_ngrok(app)  # Auto-expose the server

@app.route('/translate', methods=['POST'])
def translate():
    if 'file' not in request.files:
        return {"error": "No file uploaded"}, 400

    file = request.files['file']
    audio_path = "input.wav"
    file.save(audio_path)

    # ✅ 4. Load the Nepali speech file & resample to 16kHz if needed
    waveform, sample_rate = torchaudio.load(audio_path)

    # Resample if needed
    target_sample_rate = 16000
    if sample_rate != target_sample_rate:
        resampler = torchaudio.transforms.Resample(orig_freq=sample_rate, new_freq=target_sample_rate)
        waveform = resampler(waveform)
        print(f"Resampled audio to {target_sample_rate} Hz.")

    # ✅ 5. Process audio correctly
    inputs = processor(
        audios=[waveform.squeeze().numpy()],
        sampling_rate=target_sample_rate,
        src_lang="nep",
        return_tensors="pt"
    )

    # ✅ 6. Move `input_features` to the same device as the model
    input_features = inputs.get("input_features").to(device)

    # ✅ 7. Generate English speech output
    with torch.no_grad():
        output_audio = model.generate(input_features=input_features, tgt_lang="eng" )

    if isinstance(output_audio, tuple):
        output_audio = output_audio[0]  # Extract the audio tensor

    # ✅ 8. Save translated audio
    translated_audio_path = "translated_audio.wav"
    torchaudio.save(translated_audio_path, output_audio.cpu(), target_sample_rate)
    print(f"Translated audio saved to {translated_audio_path}")

    # ✅ 9. Return translated audio file
    return send_file(translated_audio_path, as_attachment=True, mimetype="audio/wav")

# ✅ 10. Run the Flask server
app.run()
```
